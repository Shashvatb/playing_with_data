## Importing libraries

```python
import pandas as pd
import os
```

## Reading Files

Primarily, we use 3 file formats to read data<br>
1) excel<br>
2) csv <br>
3) parquet <br>

NOTE: A lot of times data is directly read from multiple datastores, but we are talking about flat files present on the machine

### Excel file

```python
df = pd.read_excel(os.path.join(os.getcwd(), '..', 'data', 'data.xlsx'))
```

```python
df.head()
```

### CSV file - default separator

```python
df = pd.read_csv(os.path.join(os.getcwd(), '..', 'data', 'data.csv'))
```

```python
df.head()
```

### Parquet file

```python
df = pd.read_parquet(os.path.join(os.getcwd(), '..', 'data', 'data.parquet'))
```

```python
df.head()
```

### CSV file - custom separator

In this case we are using the '.txt' extension but in general, any human readble format should be fine

Note: In most cases, when tab is used as a separator, '.tsv' is used as extension

```python
df = pd.read_csv(os.path.join(os.getcwd(), '..', 'data', 'data.txt'), sep='\t')
```

```python
df.head()
```
